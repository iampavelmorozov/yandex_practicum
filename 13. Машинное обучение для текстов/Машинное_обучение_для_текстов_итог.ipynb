```python
def check(data):
    l = []
    columns = data.columns
    for col in columns:
        dtypes = data[col].dtypes
        nunique = data[col].nunique()
        sum_nan = data[col].isna().sum()
        percent_nan = data[col].isna().sum()*100/len(data)
        l.append([col, dtypes, nunique, sum_nan, percent_nan])
    data_check = pd.DataFrame(l)
    data_check.columns = ['Столбец', 'Тип данных', 'Уникальные значения', 'Сумма NaN', 'Процент NaN']
    return data_check

check(toxic).round(2)
```

## Очистка и токенизация текста

```python
corpus = toxic 
corpus.head()
```

```python
def clear_text(text):
    text = re.sub(r'(?:\n|\r)', ' ', text)
    text = re.sub(r"[^a-zA-Z ]+", ' ', text).strip()
    text = text.lower()
    return text
```

```python
corpus['cleared_text'] = corpus['text'].apply(clear_text)
corpus.head()
```

## Удаление стоп-слов

```python
stop_words = set(stopwords.words('english')) 
```

```python
def remove_stopwords(cleared_text):
        words = cleared_text.split()
        filtered_words = [word for word in words if word not in stop_words] 
        return ' '.join(filtered_words)
```

```python
corpus['no_stopwords_text'] = corpus['cleared_text'].apply(lambda x: remove_stopwords(x))
```

```python
corpus.head()
```

## POS-тегирование для назначения частей речи

```python
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {'J': wordnet.ADJ,
                'N': wordnet.NOUN,
                'V': wordnet.VERB,
                'R': wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

## Лемматизация текста

```python
def lemmatize(text):
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in nltk.word_tokenize(text)]
    lemmatized_words = ' '.join(lemmatized_words)
    
    return lemmatized_words
```

```python
corpus['lemmatized_text'] = corpus['no_stopwords_text'].progress_apply(lambda x: lemmatize(x))
corpus.head()
```

2. Целевой признак несбалансирован. Подавляющее число коментариев положительные и отлько токсичные комментарии составляют 10.16 процента (16186 комментариев). Учтем несбалансированность классов в целевом признаке при построении моделей машинного обучения.
